```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("1. Weather Data.csv")
```

```python
data.head()
```

```python
# 1. Find all records where the weather was exactly clear.
# Filter records where the weather is exactly 'Clear'
clear_weather_records = data[data['Weather'] == 'Clear']
# Display the filtered records
print(clear_weather_records)
```

```python
# 2.Find the number of times the wind speed was exactly 4 km/hr.
# Count the records where the wind speed is exactly 4 km/hr
wind_speed_count = data[data['Wind Speed_km/h'] == 4].shape[0]
print(f"The wind speed was exactly 4 km/hr {wind_speed_count} times.")
```

```python
# 3. Check if there are any NULL values present in the dataset.
# Check for NULL values in each column
null_values = data.isnull().sum()

# Display the result
print(null_values)
```

```python
# 4. Rename the column "Weather" to "Weather_Condition"
data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)

# Display the DataFrame to verify the change
print(data.head())
```

```python
# 5. What is the mean visibility of the dataset?
# Calculate the mean visibility
mean_visibility = data['Visibility_km'].mean()

# Display the mean visibility
print(f"The mean visibility is: {mean_visibility}")
```

```python
# 6. Filter records where wind speed is greater than 24 km/hr and visibility is equal to 25 km
filtered_records = data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]

# Count the number of such records
num_records = filtered_records.shape[0]

# Display the count
print(f"Number of records where wind speed is greater than 24 km/hr and visibility is 25 km: {num_records}")
```

```python
# 7. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
# Group the data by the 'Weather' column and calculate the mean for each group
# Select only numeric columns for calculating the mean
numeric_data = data.select_dtypes(include='number')

# Add the 'Weather' column back to the numeric data for grouping
numeric_data['Weather_Condition'] = data['Weather_Condition']

# Group the data by the 'Weather' column and calculate the mean for each group
mean_values = numeric_data.groupby('Weather_Condition').mean()

# Display the mean values for each weather condition
print(mean_values)
```

```python
# 8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
# Filter records based on the conditions
filtered_records = data[
    (data['Weather_Condition'] == 'Clear') &
    (data['Rel Hum_%'] > 50) |
    (data['Visibility_km'] > 40)
]

# Display the filtered records
print(filtered_records)
```

```python
# 9. Find the number of weather conditions that include snow.
# Filter weather conditions that include the word "Snow"
snow_conditions = data['Weather_Condition'].str.contains('Snow', case=False, na=False)

# Find the unique weather conditions that include "Snow"
unique_snow_conditions = data[snow_conditions]['Weather_Condition'].unique()

# Count the number of unique weather conditions that include "Snow"
num_snow_conditions = len(unique_snow_conditions)

# Display the count
print(f"Number of distinct weather conditions that include 'Snow': {num_snow_conditions}")
```
